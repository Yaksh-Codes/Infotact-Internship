# Import Libraries

```python
# Data Manipulation
import pandas as pd
import numpy as np

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Settings
pd.set_option('display.max_columns', None)
sns.set_style("whitegrid")
```

# Load Datasets

```python
# Load datasets

marketing_df = pd.read_csv("Marketing.csv")

customers_df = pd.read_csv("olist_customers_dataset.csv")

retail_df = pd.read_excel("Online Retail.xlsx")

print("Marketing Shape:", marketing_df.shape)
print("Customers Shape:", customers_df.shape)
print("Retail Shape:", retail_df.shape)
```

# Basic Data Inspection

```python
datasets = {
    "Marketing": marketing_df,
    "Customers": customers_df,
    "Retail": retail_df
}

for name, df in datasets.items():
    print(f"\n{name} Dataset")
    print("="*50)

    print("\nFirst 5 Rows")
    display(df.head())

    print("\nInfo")
    print(df.info())

    print("\nMissing Values")
    print(df.isnull().sum())

    print("\nDuplicate Rows")
    print(df.duplicated().sum())
```

# Handle Missing UTM Parameters

```python
utm_cols = ['utm_source', 'utm_medium', 'utm_campaign']

for col in utm_cols:
    if col in marketing_df.columns:
        marketing_df[col] = marketing_df[col].fillna('Unknown')

print("UTM Missing Values Handled")
```

# Standardize Timestamp Formats

Marketing Dataset

```python
if 'c_date' in marketing_df.columns:
    marketing_df['c_date'] = pd.to_datetime(
        marketing_df['c_date'],
        errors='coerce',
        utc=True
    )

marketing_df['c_date'].head()
```

Online Retail Dataset

```python
retail_df['InvoiceDate'] = pd.to_datetime(
    retail_df['InvoiceDate'],
    errors='coerce',
    utc=True
)

retail_df['InvoiceDate'].head()
```

# Remove Duplicate User IDs

Customers Dataset

```python
print("Original Shape:", customers_df.shape)

customers_df = customers_df.drop_duplicates(
    subset=['customer_unique_id']
)

print("After Deduplication:", customers_df.shape)
```

Marketing Dataset

```python
if 'id' in marketing_df.columns:
    marketing_df = marketing_df.drop_duplicates(subset=['id'])
```

# Missing Value Analysis

```python
missing_df = pd.DataFrame({
    'Missing Values': marketing_df.isnull().sum(),
    'Percentage': (
        marketing_df.isnull().sum() /
        len(marketing_df)
    ) * 100
})

missing_df.sort_values(
    by='Percentage',
    ascending=False
)
```

# Campaign Performance Analysis

```python
campaign_analysis = marketing_df.groupby(
    'campaign_name'
).agg({
    'impressions':'sum',
    'clicks':'sum',
    'leads':'sum',
    'orders':'sum',
    'revenue':'sum',
    'mark_spent':'sum'
}).reset_index()

campaign_analysis.head()
```

# CTR (Click Through Rate)

```python
marketing_df['CTR'] = (
    marketing_df['clicks'] /
    marketing_df['impressions']
) * 100

marketing_df[['campaign_name','CTR']].head()
```

# Conversion Rate

```python
marketing_df['Conversion_Rate'] = (
    marketing_df['orders'] /
    marketing_df['clicks']
) * 100

marketing_df[['campaign_name','Conversion_Rate']].head()
```

# ROAS (Return on Ad Spend)

```python
marketing_df['ROAS'] = (
    marketing_df['revenue'] /
    marketing_df['mark_spent']
)

marketing_df[['campaign_name','ROAS']].head()
```

# Top Campaigns by Revenue

```python
top_campaigns = marketing_df.groupby(
    'campaign_name'
)['revenue'].sum().sort_values(
    ascending=False
).head(10)

plt.figure(figsize=(12,6))
top_campaigns.plot(kind='bar')
plt.title("Top Campaigns by Revenue")
plt.ylabel("Revenue")
plt.show()
```

# Revenue Distribution

```python
plt.figure(figsize=(10,6))

sns.histplot(
    marketing_df['revenue'],
    bins=30,
    kde=True
)

plt.title("Revenue Distribution")
plt.show()
```

# Correlation Heatmap

```python
numeric_cols = marketing_df.select_dtypes(
    include=np.number
)

plt.figure(figsize=(12,8))

sns.heatmap(
    numeric_cols.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Matrix")
plt.show()
```

# Country-wise Sales Analysis

```python
country_sales = retail_df.groupby(
    'Country'
)['Quantity'].sum().sort_values(
    ascending=False
).head(10)

plt.figure(figsize=(12,6))

country_sales.plot(kind='bar')

plt.title("Top Countries by Quantity Sold")
plt.ylabel("Quantity")
plt.show()
```

# Customer Distribution by State

```python
customers_df['customer_state'].value_counts().head(10).plot(
    kind='bar',
    figsize=(10,6)
)

plt.title("Top States by Customers")
plt.show()
```

# Data Quality Report

```python
def data_quality_report(df):

    report = pd.DataFrame({
        'Data Type': df.dtypes,
        'Missing Values': df.isnull().sum(),
        'Duplicate Count': df.duplicated().sum()
    })

    return report

data_quality_report(marketing_df)
```
